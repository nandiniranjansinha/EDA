```python
# Will change Date_of_Journey to int
# We can convert to Date format as well but ultimately it has to be converted to int
# In case of forecasting, we can keep dtype as Date
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```
